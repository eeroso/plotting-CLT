```
import collections
import numpy
import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions

try:
  tf.compat.v1.enable_eager_execution()
except ValueError:
  pass

import matplotlib.pyplot as plt

b = tfd.Bernoulli(probs=0.6)
mean10_ber = []
mean20_ber = []
mean100_ber = []

n = tfd.Normal(loc=0., scale=1/10.)
n20 = tfd.Normal(loc=0., scale=1/20.)
n100 = tfd.Normal(loc=0., scale=1/100.)
mean10 = []
mean20 = []
mean100 = []

u = tfd.Uniform(low=0.0, high=1.0)
mean10_uni = []
mean20_uni = []
mean100_uni = []

batch_sz_1 = 10
batch_sz_2 = 20
batch_sz_3 = 100

for _ in range(100000):
  sample = n.sample(10)
  sample_b = b.sample(10)
  sample_u = u.sample(10)
  mean = numpy.mean(sample)
  mean_b = numpy.mean(sample_b)
  mean_u = numpy.mean(sample_u)
  mean10.append(mean)
  mean10_ber.append(mean_b)
  mean10_uni.append(mean_u)

  sample20 = n20.sample(20)
  sample20_b = b.sample(20)
  sample20_u = u.sample(20)
  mean20_b = numpy.mean(sample20_b)
  mean20_u = numpy.mean(sample20_u)
  mean20_ = numpy.mean(sample20)
  mean20.append(mean20_)
  mean20_ber.append(mean20_b)
  mean20_uni.append(mean20_u)

  
  sample100 = n100.sample(100)
  sample100_b = b.sample(100)
  sample100_u = u.sample(100)

  mean100_b = numpy.mean(sample100_b)
  mean100_u = numpy.mean(sample100_u) 
  mean100_ = numpy.mean(sample100)
  mean100.append(mean100_)
  mean100_ber.append(mean100_b)
  mean100_uni.append(mean100_u)



_ = plt.hist(mean10, bins='auto')
plt.title("Histogram for batch size 10 normal dis")
plt.show()

_ = plt.hist(mean20, bins='auto')
plt.title("Histogram for batch size 20 normal dis")
plt.show()

_ = plt.hist(mean100, bins='auto')
plt.title("Histogram for batch size 100 normal dis")
plt.show()

_ = plt.hist(mean10_ber, bins='auto')
plt.title("Histogram for batch size 10 bernoille dis")
plt.show()

_ = plt.hist(mean20_ber, bins='auto')
plt.title("Histogram for batch size 20 bernoille dis")
plt.show()

_ = plt.hist(mean100_ber, bins='auto')
plt.title("Histogram for batch size 100 bernoille dis")
plt.show()

_ = plt.hist(mean10_uni, bins='auto')
plt.title("Histogram for batch size 10 uniform dis")
plt.show()

_ = plt.hist(mean20_uni, bins='auto')
plt.title("Histogram for batch size 20 uniform dis")
plt.show()

_ = plt.hist(mean100_uni, bins='auto')
plt.title("Histogram for batch size 100 uniform dis")
plt.show()



  
  
```
